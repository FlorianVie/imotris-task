```python
import numpy as np
import pandas as pd
import random

ingr = ["Salade", "Tomate", "Oignon", "Fromage", "Viande", "Poivron"]
nbIngr = 6
totIngr = 14
maxIngr = 5
intCons = [14, 14]
nbDist = 1000
```

```python
distributions = []

while len(distributions) < nbDist:
    gen = np.random.randint(maxIngr + 1, size=nbIngr)
    if gen.sum() >= intCons[0] and gen.sum() <= intCons[1]:
        distributions.append(gen)

dist_table = pd.DataFrame(distributions, columns=ingr)
dist_table.to_csv("ingredients.csv", index=False)
dist_table
```
